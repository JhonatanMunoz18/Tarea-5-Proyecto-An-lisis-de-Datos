### 3. Seleccionar las características más relevantes para entrenar el 
### modelo utilizando selección de características.

```python
# Seleccionar las columnas deseadas para entrenar
data_seleccionado = data[['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
```

```python
# Ver el resultado
print(data_seleccionado.head())
```

```python
#Total de perssonas que sobrovivieron
data_seleccionado['Survived'].value_counts()
```

### 4. Dividir el dataset en Train y Test para evaluar correctamente el 
### modelo. 

```python
# Porcentaje de datos para entrenamiento (train) = 75% , y para prueba (test) = 25%

# Se crean variables X y Y para almacenar la información del entrenamiento

X = data_seleccionado.drop('Survived', axis=1) # Se cargan los datos de train sin los datos de la columna sobrevivientes
Y = data_seleccionado['Survived' ] # Se cargan Los datos de train con los datos de la la columna sobrevivientes

#La función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state= 0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

### 5. Entrenar el modelo configurando los diferentes hiperparámetros. 

```python
# Construir el modelo de regresión logística

modelo_regresion = LogisticRegression(solver = "liblinear") # liblinear es una libreria para resolver problemas lineales de clasificacion y regresion
modelo_regresion.fit (X_train , Y_train)

# Crea el modelo de prediccion
Prediccion = modelo_regresion.predict(X_test) # la función predict predice los valores a partir del conjunto de datos de validación
```

```python
print(Y_test)
```

```python
print(Prediccion)
```

### 6. Evaluar el desempeño del modelo en el conjunto de Test con 
### métricas como precisión, recall, F1-score, etc. 

```python
# Desempeño del modelo de regresión

print(classification_report(Y_test,Prediccion))
```

### 7. Realizar las diferentes gráficas que permitan visualizar los 
### resultados del modelo. 


```python
#Matrix de Confusión del modelo

cm=confusion_matrix(Y_test, Prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Curva de precisón del modelo

from sklearn.metrics import RocCurveDisplay
RocCurveDisplay. from_predictions(Y_test, Prediccion)
plt.show()
```
